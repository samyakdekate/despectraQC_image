```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import os




# Step 1: Load Image or generate a fallback
def load_image(image_path):
    if not os.path.exists(image_path):
        print(f"⚠️ File '{image_path}' not found. Generating a random test image instead...")
        arr = np.random.randint(0, 256, (100, 100), dtype=np.uint8)  # 100x100 grayscale
        img = Image.fromarray(arr)
        img.save("generated_test_image.jpg")
        image_path = "generated_test_image.jpg"

    img = Image.open(image_path).convert("L")  # grayscale
    width, height = img.size
    total_pixels = width * height
    print(f"✅ Using image: {image_path}")
    print(f"Image size: {width} x {height}")
    print(f"Total pixels: {total_pixels}")
    return img, total_pixels

# Step 2: Quantum-inspired image contrast
def quantum_contrast(img):
    arr = np.array(img)
    processed = np.zeros_like(arr)

    # Iterate pixels (simulate encoding pixel into qubit)
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            brightness = arr[i, j] / 255.0  # normalize

            # Encode brightness as probability amplitude
            qc = QuantumCircuit(1)
            qc.ry(brightness * np.pi, 0)  # rotation encodes pixel

            # Apply a gate for "quantum contrast"
            qc.x(0)  # X gate inverts the state

            # Directly get statevector
            state = Statevector.from_instruction(qc)

            # Probability of |1>
            prob_one = abs(state.data[1]) ** 2

            # Map back to 0–255
            processed[i, j] = int(prob_one * 255)

    return Image.fromarray(processed)

# Step 3: Compare images & show stats
def compare_images(img1, img2):
    arr1 = np.array(img1)
    arr2 = np.array(img2)

    stats = {
        "Original Avg Brightness": np.mean(arr1),
        "Quantum Avg Brightness": np.mean(arr2),
        "Original Min Pixel": np.min(arr1),
        "Quantum Min Pixel": np.min(arr2),
        "Original Max Pixel": np.max(arr1),
        "Quantum Max Pixel": np.max(arr2)
    }

    print("\n📊 Comparison Details:")
    for k, v in stats.items():
        print(f"{k}: {v:.2f}")

    return stats

# Step 4: Main analysis
def analyze_image(image_path):
    img, total_pixels = load_image(image_path)
    q_img = quantum_contrast(img)

    # Show side by side
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(img, cmap='gray')
    axes[0].set_title("Original Image")
    axes[0].axis("off")

    axes[1].imshow(q_img, cmap='gray')
    axes[1].set_title("Quantum Enhanced Image")
    axes[1].axis("off")

    plt.show()

    # Print stats
    compare_images(img, q_img)

# Run Example
if __name__ == "__main__":
    analyze_image("sample_image.jpg")
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# Step 1: Load Image or generate a fallback
def load_image(image_path):
    if not os.path.exists(image_path):
        print(f"⚠️ File '{image_path}' not found. Generating a random test image instead...")
        arr = np.random.randint(0, 256, (100, 100), dtype=np.uint8)  # 100x100 grayscale
        img = Image.fromarray(arr)
        img.save("generated_test_image.jpg")
        image_path = "generated_test_image.jpg"

    img = Image.open(image_path).convert("L")  # grayscale
    width, height = img.size
    total_pixels = width * height
    print(f"✅ Using image: {image_path}")
    print(f"Image size: {width} x {height}")
    print(f"Total pixels: {total_pixels}")
    return img, total_pixels

# ======================
# Quantum Edge Detector
# ======================

class QuantumEdgeDetector:
    def __init__(self):
        self.simulator = AerSimulator()

    def encode_pixels(self, p1, p2):
        """Encode two neighboring pixel values into a quantum circuit"""
        qc = QuantumCircuit(2, 2)
        angle1 = (p1 / 255.0) * np.pi
        angle2 = (p2 / 255.0) * np.pi

        qc.ry(angle1, 0)
        qc.ry(angle2, 1)
        qc.cx(0, 1)   # entanglement
        qc.h(0)       # superposition
        qc.measure([0, 1], [0, 1])
        return qc

    def run_circuit(self, qc, shots=512):
        job = self.simulator.run(qc, shots=shots)
        result = job.result()
        counts = result.get_counts()
        return counts.get('11', 0) / shots  # Probability of "11"

    def detect_edges(self, image, step=1):
        """Perform quantum edge detection"""
        h, w = image.shape
        edge_map = np.zeros_like(image, dtype=float)

        # Compare right and bottom neighbors (like Sobel operator)
        for i in range(h - step):
            for j in range(w - step):
                p1 = image[i, j]
                p2 = image[i, j + step]   # horizontal neighbor
                p3 = image[i + step, j]   # vertical neighbor

                # Quantum comparisons
                qc1 = self.encode_pixels(p1, p2)
                qc2 = self.encode_pixels(p1, p3)

                val_h = self.run_circuit(qc1)
                val_v = self.run_circuit(qc2)

                # Combine like Sobel gradient magnitude
                edge_strength = np.sqrt(val_h**2 + val_v**2)
                edge_map[i, j] = edge_strength

        # Normalize to 0–255 for high contrast
        edge_map = (edge_map / np.max(edge_map)) * 255
        return edge_map.astype(np.uint8)

if __name__ == "__main__":
    # Load grayscale image
    path = "C:/Users/Samyak/Desktop/MINI_QC/sample_image.jpg"
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (64, 64))  # keep small for quantum simulation

    detector = QuantumEdgeDetector()
    q_edges = detector.detect_edges(img, step=1)

    # Classical Sobel for comparison
    sobel_edges = cv2.Sobel(img, cv2.CV_64F, 1, 1, ksize=3)
    sobel_edges = cv2.convertScaleAbs(sobel_edges)

    # Show results
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1), plt.imshow(img, cmap='gray'), plt.title("Original"), plt.axis("off")
    plt.subplot(1,3,2), plt.imshow(q_edges, cmap='gray'), plt.title("Quantum Edge Detection"), plt.axis("off")
    plt.subplot(1,3,3), plt.imshow(sobel_edges, cmap='gray'), plt.title("Classical Sobel"), plt.axis("off")
    plt.show()
# ======================
# Analysis
# ======================

# Step 2: Quantum-inspired image contrast
def quantum_contrast(img):
    arr = np.array(img)
    processed = np.zeros_like(arr)

    # Iterate pixels (simulate encoding pixel into qubit)
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            brightness = arr[i, j] / 255.0  # normalize

            # Encode brightness as probability amplitude
            qc = QuantumCircuit(1)
            qc.ry(brightness * np.pi, 0)  # rotation encodes pixel

            # Apply a gate for "quantum contrast"
            qc.x(0)  # X gate inverts the state

            # Directly get statevector
            state = Statevector.from_instruction(qc)

            # Probability of |1>
            prob_one = abs(state.data[1]) ** 2

            # Map back to 0–255
            processed[i, j] = int(prob_one * 255)

    return Image.fromarray(processed)

# Step 3: Compare images & show stats
def compare_images(img1, img2):
    arr1 = np.array(img1)
    arr2 = np.array(img2)

    diff = np.abs(arr1.astype(int) - arr2.astype(int))

    stats = {
        "Original Avg Brightness": np.mean(arr1),
        "Quantum Avg Brightness": np.mean(arr2),
        "Original Min Pixel": np.min(arr1),
        "Quantum Min Pixel": np.min(arr2),
        "Original Max Pixel": np.max(arr1),
        "Quantum Max Pixel": np.max(arr2),
        "Mean Pixel Difference": np.mean(diff),
        "Max Pixel Difference": np.max(diff),
    }

    print("\n📊 Comparison Details:")
    for k, v in stats.items():
        print(f"{k}: {v:.2f}")

    return stats, diff

# Step 4: Main analysis
def analyze_image(image_path):
    img, total_pixels = load_image(image_path)
    q_img = quantum_contrast(img)

    # Compare images
    stats, diff = compare_images(img, q_img)

    # Show side by side (4 plots)
    fig, axes = plt.subplots(1, 4, figsize=(20, 5))

    axes[0].imshow(img, cmap='gray')
    axes[0].set_title("Original Image")
    axes[0].axis("off")

    axes[1].imshow(q_img, cmap='gray')
    axes[1].set_title("Quantum Enhanced Image")
    axes[1].axis("off")

    im = axes[2].imshow(diff, cmap='inferno')
    axes[2].set_title("Pixel Difference Heatmap")
    axes[2].axis("off")
    fig.colorbar(im, ax=axes[2], fraction=0.046, pad=0.04)

    # Text box with comparison stats
    stats_text = "\n".join([f"{k}: {v:.2f}" for k, v in stats.items()])
    axes[3].axis("off")
    axes[3].text(0.1, 0.5, stats_text, fontsize=12, verticalalignment="center")

    plt.tight_layout()
    plt.show()

# Run Example
if __name__ == "__main__":
    analyze_image("sample_image.jpg")
```
